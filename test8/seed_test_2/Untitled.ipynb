```python
import cv2
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline  
import tensorflow as tf
from glob import glob
from yolov3.yolov4_config import Create_Yolo
from yolov3.utils_config import load_yolo_weights, detect_image
from yolov3.configs_config import *
from PIL import Image
import random

import os
import pandas as pd

if YOLO_TYPE == "yolov4":
    Darknet_weights = YOLO_V4_TINY_WEIGHTS if TRAIN_YOLO_TINY else YOLO_V4_WEIGHTS
if YOLO_TYPE == "yolov3":
    Darknet_weights = YOLO_V3_TINY_WEIGHTS if TRAIN_YOLO_TINY else YOLO_V3_WEIGHTS
```

```python
yolo = Create_Yolo(input_size=YOLO_INPUT_SIZE, training = True)
load_yolo_weights(yolo, Darknet_weights) # use Darknet weights
```

```python
yolo.summary()
```

```python
print("the number of layers",len(yolo.get_weights()))
l = 0
for i in yolo.get_weights():
    l +=sum(i.reshape([-1]))
print("all weight sum is : ",l)
```
